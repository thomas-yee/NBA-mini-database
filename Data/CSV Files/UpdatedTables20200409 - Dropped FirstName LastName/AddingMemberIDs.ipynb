# Inserting MemberIDs into member csv file

```python
import csv
```

```python
def insertID(inputFileName, outputFileName):
    with open(outputFileName, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        with open(inputFileName, 'r') as csvfile:
            reader = csv.reader(csvfile)
            headerLine = True
            playerID = 0
            for row in reader:
                if headerLine:
                    headers = ['MemberID']
                    for attribute in row:
                        headers.append(attribute)
                    writer.writerow(headers)
                    print(headers)
                    headerLine = False
                    continue
                newRow = [playerID]
                playerID += 1
                for attribute in row:
                    newRow.append(attribute)
                print(newRow)
                writer.writerow(newRow)
    writer.close()
    reader.close()
```

```python
insertID('member.csv', 'memberUpdated.csv')
```

# Inserting and matching MemberIDs from memberUpdated.csv into tables: Coach, CoachStats, Player, PlayerStats

```python
import pandas as pd
```

```python
memberdf = pd.read_csv("memberUpdated.csv")
memberdf
```

```python
def getMemberID(firstName, lastName):
    memberID = memberdf.loc[((memberdf['FirstName'] == firstName) & (memberdf['LastName'] == lastName))]['MemberID'].values[0]
    return memberID
```

```python
def insertMemberID(inputFileName, outputFileName):
    with open(outputFileName, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        with open(inputFileName, 'r') as csvfile:
            reader = csv.reader(csvfile)
            headerLine = True
            for row in reader:
                try:
                    if headerLine:
                        headers = ['MemberID']
                        count = 0
                        for attribute in row:
                            count += 1
                            if count < 3:
                                continue
                            headers.append(attribute)
                        writer.writerow(headers)
                        print(headers)
                        headerLine = False
                        continue
                    memberID = [getMemberID(row[0], row[1])]
                    oldRow = []
                    count = 0
                    for attribute in row:
                        count += 1
                        if count < 3:
                            continue
                        oldRow.append(attribute)
                    newRow = memberID + oldRow
                    print(newRow)
                    writer.writerow(newRow)
                except:
                    continue
```

```python
insertMemberID('player_stats.csv', 'player_statsUpdated.csv')
```

# Fixing PPG in player_stats.csv

```python
pstat_properPPG = pd.read_csv("player_stats_properPPG.csv")
pstat_properPPG
```

```python
dateCorrections = {'97':'1997', '98':'1998', '99':'1999', '00':'2000', '01':'2001', '02':'2002', '03':'2003', '04':'2004', '05':'2005', '06':'2006', '07':'2007', '08':'2008','09':'2009', '10':'2010', '11':'2011', '12':'2012', '13':'2013', '14':'2014', '15':'2015', '16':'2016', '17':'2017', '18':'2018', '19':'2019', '20':'2020',
                   'Jan':'2001', 'Feb':'2002', 'Mar':'2003', 'Apr':'2004', 'May':'2005', 'Jun':'2006', 'Jul':'2007', 'Aug':'2008', 'Sep':'2009', 'Oct':'2010', 'Nov':'2011', 'Dec':'2012'}
```

```python
def mapDate(unformatted):
    try:
        mappedDate = dateCorrections[unformatted]
    except:
        mappedDate = unformatted
    return mappedDate
```

```python
print(mapDate('Jan'))
```

```python
def fixPlayerStatsTable(inputFileName, outputFileName):
    with open(outputFileName, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        with open(inputFileName, 'r') as csvfile:
            reader = csv.reader(csvfile)
            skipHeader = True
            for row in reader:
                if skipHeader:
                    writer.writerow(row)
                    skipHeader = False
                    continue
                date = row[2].split('-')
                date[0] = mapDate(date[0])
                date[1] = mapDate(date[1])
                date = date[0] + '-' + date[1]
                newRow = row[0:2]
                newRow.append(date)
                for attribute in row[3:]:
                    newRow.append(attribute)
                writer.writerow(newRow)
```

```python
fixPlayerStatsTable('player_stats_properPPG.csv', 'player_stats_correctedPPG.csv')
```

# Insert decimal to PPG if whole number

```python
def fixPPGTenth(inputFileName, outputFileName):
    with open(outputFileName, 'w', newline='') as csv_file:
        writer = csv.writer(csv_file, delimiter=',')
        with open(inputFileName, 'r') as csvfile:
            reader = csv.reader(csvfile)
            skipHeader = True
            for row in reader:
                if skipHeader:
                    writer.writerow(row)
                    skipHeader = False
                    continue
                ppg = row[-1]
                if ppg == '':
                    continue
                elif len(ppg.split('.')) == 1:
                    ppg = str(ppg) + '.0'
                newRow = row[0:-1]
                newRow.append(ppg)
                writer.writerow(newRow)
```

```python
fixPPGTenth('player_statsNoTenth.csv', 'player_statsUpdated.csv')
```
